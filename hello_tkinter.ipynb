# Intro


````python
# import tkinter
from tkinter import *

# create a root widget ie the boxy window thibnng
root = Tk()

# create a label widget
myLabel = Label(root, text="Hello, Ben? Is that you?")

# Shove the program into the window, best fit
myLabel.pack()

# finally, create an event loop to track events
root.mainloop()
````

* `from tkinter import *`: This line imports all the classes, functions, and constants from the tkinter module. The asterisk (*) means we import everything, so we don't need to specify each item explicitly.

* `root = Tk()`: This line creates the main window of the application. The Tk() function is used to instantiate the Tkinter class, which represents the root or main window of the application.

* `myLabel = Label(root, text="Hello, Ben? Is that you?")`: This line creates a Label widget, which is a text display area, and assigns it to the variable myLabel. The Label widget is initialized with the root window as its parent and the text "Hello, Ben? Is that you?" as its content.

* `myLabel.pack()`: This line uses the pack() method to insert the myLabel widget into the root window. The pack() method organizes widgets in a block-like manner, automatically adjusting their size and position.

* `root.mainloop()`: This line starts the Tkinter event loop. The mainloop() method is responsible for handling user interactions, such as button clicks or window resizing. It continuously listens for events and updates the GUI accordingly until the window is closed.

# Grid system

The grid() method in Tkinter allows you to create a grid-based layout for organizing widgets within a window or frame. It provides a more flexible and precise way to position widgets compared to the pack() method.

When using the grid() method, you can specify the row and column positions for each widget in the grid. The grid uses a coordinate system where the top-left cell has coordinates (0, 0), and each cell is identified by its row and column index.

````python
# create a root widget ie the boxy window thibnng
root = Tk()

# create a label widget
myLabel_1 = Label(root, text="Hello, Ben? Is that you?")
myLabel_2 = Label(root, text="Yeah, Its me?")

# Shove the program into the window, best fit
myLabel_1.grid(row=0, column=0)
myLabel_2.grid(row=1, column=0)

# finally, create an event loop to track events
root.mainloop()
````

'`normal`' (default): The button is in the normal state and can be clicked and interacted with.

'`disabled`': The button is disabled and cannot be clicked or interacted with. It appears grayed out to indicate its disabled state.

## storing the numbers clicked as global variables

in order to perform arithmetic operations on the numbers entered, the program has to somehow be able to remember the previous number entered before the operator.

eg if we enter 2+2=? the process should be
enter 2
enter + which clears 2 and at the same time stores 2
enter 2 which clears + and at the same time stores +
enter = which clears 2 and at the same time stores 2
get result which is 4

## Next, how will the equal button know which arithmetic operand is clicked?

as we have seen, the `button_add` function works . so we will define new functions for multiply, divide and subtract and let the equal function decide which sign was entered. to notify the equal sign we will use a global variable `math` that stores the characteristic of the function such that if the button is presssed, the variable sent to the equal sign will be an arithmetic operations identifier.
